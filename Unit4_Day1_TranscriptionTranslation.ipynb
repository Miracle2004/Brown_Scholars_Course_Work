Now that we've assigned our sequences to variables, we can get started on the code to transcribe that DNA sequence into an RNA sequence. 

**Store the histamine receptor RNA sequence in its own variable and print it out to check your results**

Remember, A -> U, T -> A, G -> C, and C -> G

```python
# Hint: You will need a "for loop" and conditionals. 
# You will also need an empty string
Transcription = {'A':'U', 'T': 'A', 'G':'C', 'C': 'G'} #disctionary assigns the corresponding rna nucleotides to their dna pairs
HisRNA = ""
for i in HisDNA:
  if i == 'A':
    HisRNA += Transcription['A']
  elif i == 'T':
    HisRNA += Transcription['T']
  elif i =='G':
    HisRNA += Transcription['G']
  elif i == 'C':
   HisRNA += Transcription['C']
```

```python
histProtein = ""
codon = "" #all new values will go here
for i in HisRNA: #loops through the RNA list
  if len(codon) == 3: #limits length of codon
    histProtein += Translation[codon] #uses the 3 values stored in the variable and checks for it in the values of the dictionary then adds it to histProtein
    codon = "" #empties out the codon holder
    codon += i #ensures we don't skip the character in the next iteration
  else:
    codon += i #checks if the length of codon is 3 and if it isn't it will add another character until it is
print(histProtein)
```
