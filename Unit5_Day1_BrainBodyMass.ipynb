## Making a function

Now it's time to make our function! You want to write a function that takes the name of the dataframe and the name of two columns and returns the average brain to body mass ratio. 

Remember: a brain to body mass ratio is simply the brain mass divided by the body mass

It's probably a good idea to start with pseudocode!

```python
def braintobody(df_name, column_name1, column_name2): #function header
  ratio = df_name[column_name1]/df_name[column_name2] #brain mass/weight to body mass/weight ratios stored in another variable
  average = ratio.mean() #store the mean of the data into another variable

  return average
```

```python
average_mammal = braintobody(df_mammals, "Brain mass (g)", "Body mass (g)")
print(average_mammal)
```
