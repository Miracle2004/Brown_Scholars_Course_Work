```python
# ornithischian: 10**(3.0587*np.log10(FL)-2.7042)/1000

# sauropod: 10**(2.3459*np.log10(FL)-0.2935)/1000

# theropod: 10**(3.1854*np.log10(FL)-3.184)/1000
```

Using these formulas, create a function that takes *two arguments:* the **dinosaur type** and the **femur length in mm** and *returns* the **dinosaur's body mass**

```python
def bodyMassCalculation(clade, Length):
  if clade == "ornithischian":
    bodyMass = 10**(3.0587*np.log10(Length)-2.7042)/1000
  elif clade == "sauropod":
    bodyMass = 10**(2.3459*np.log10(Length)-0.2935)/1000
  elif clade == "theropod":
    bodyMass = 10**(3.1854*np.log10(Length)-3.184)/1000
  return bodyMass
```

Next, create a function that takes *one argument:* the **dinosaur type** and *returns* the **average body mass** for that dinosaur type.

(Note: You do not need to specify the data frame as an argument, because we are only working with a single data frame)

Hints: 
- You will need to use the .loc[] index command that we learned for the Astrophysics unit project.
- The three dinosaur types (Clades) in this data set are: Ornithischia, Sauropodomorpha, and Theropoda

```python
# ornithischian: 10**(3.0587*np.log10(FL)-2.7042)/1000

# sauropod: 10**(2.3459*np.log10(FL)-0.2935)/1000

# theropod: 10**(3.1854*np.log10(FL)-3.184)/1000
```

```python
def massCalc(dino_type):
  if dino_type == "Ornithischia":
    FL = dino_measurement["Femur length (mm)"].loc[dino_measurement["Clade"] =="Ornithischia"]
    bodyM = 10**(3.0587*np.log10(FL)-2.7042)/1000
    avg_bodyM = bodyM.mean()
  elif dino_type == "Sauropodomorpha":
    FL = dino_measurement["Femur length (mm)"].loc[dino_measurement["Clade"] == "Sauropodomorpha"]
    bodyM = 10**(2.3459*np.log10(FL)-0.2935)/1000
    avg_bodyM = bodyM.mean()
  elif dino_type == "Theropoda":
    FL = dino_measurement["Femur length (mm)"].loc[dino_measurement["Clade"] == "Theropoda"]
    bodyM = 10**(3.1854*np.log10(FL)-3.184)/1000
    avg_bodyM = bodyM.mean()
  return avg_bodyM
```

Test out your function! Call it for each dinosaur type and print the result:

```python
ornithischian = massCalc("Ornithischia")
print(ornithischian)
```

```python
sauropod = massCalc("Sauropodomorpha")
print(sauropod)
```

```python
theropod = massCalc("Theropoda")
print(theropod)
```
